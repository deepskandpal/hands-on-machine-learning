```
import pandas as pd
import os
import numpy as np
DIR = os.getcwd()
DIR
!ls
```

```

train_file = os.path.join(DIR, "sample_data", "california_housing_train.csv")
train_data = pd.read_csv(train_file)
train_data.head()
```

```
train_data.describe()
```

## Exploratory Data Analysis

```
import matplotlib.pyplot as plt
%matplotlib inline
train_data.hist(bins=50, figsize=(20,15))
plt.show()
```

```
train_data.plot(kind="scatter", x="longitude", y="latitude")
```

```
train_data.plot(kind="scatter", x="longitude", y="latitude", alpha=0.1)
```

```
train_data.plot(kind='scatter',x='longitude', y='latitude', alpha=0.4,s=train_data["population"]/100, label='population', figsize=(10,7),c='median_house_value',cmap=plt.get_cmap("jet"),colorbar=True)
plt.legend()
```

## Correlations

```
corr_matrix = train_data.corr()
corr_matrix['median_house_value'].sort_values(ascending=False)
```

```
from pandas.plotting import scatter_matrix
attributes = ['median_house_value','median_income','total_rooms','housing_median_age']
scatter_matrix(train_data[attributes], figsize=(12,8))
```
